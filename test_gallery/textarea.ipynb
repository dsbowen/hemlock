```python
from hemlock import create_test_app
from hemlock.questions import Textarea

app = create_test_app()
```

```python
Textarea("Write a short essay.").display()
```

```python
Textarea(floating_label="Comment").display()
```

```python
Textarea("This is a higher textarea.", textarea_tag={"style": {"height": "200px"}}).display()
```

```python
Textarea(
    """
    This textarea has a minimum number of characters and a maximum number of words.

    Note that the minimum number of characters and words means
    *the minimum number of characters or words given that the user wrote anything at all.*

    So, if the user doesn't respond, this won't be flagged by `minlength` or `minwords`.
    """,
    textarea_tag={
        "minlength": 10,
        "maxwords": 20
    }
).display()
```

```python
textarea = Textarea("Write an essay.")
textarea.set_is_valid(False)
textarea.feedback = "Your essay wasn't good enough."
textarea.display()
```

```python
textarea = Textarea(
    "Which is better and why: Hemlock or Qualtrics?",
    default="Hemlock is better than Qualtrics for many reasons, including...",
    floating_label="Essay",
    prepend=["a", "b", "c"],
    append=["x", "y", "z"],
    form_text="Here is some form text.",
    textarea_tag={
        "maxlength": 100,
        "minwords": 5,
        "maxwords": 30
    }
)
textarea.set_is_valid(True)
textarea.feedback = "Nice answer!"
textarea.display()
```
