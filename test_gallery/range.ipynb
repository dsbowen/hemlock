```python
from hemlock import create_test_app
from hemlock.questions import Range

app = create_test_app()
```

```python
Range("Pick a number.").display()
```

```python
Range(
    "Pick a number between 1 and 10.",
    input_tag={"min": 0, "max": 10, "step": .5}
).display()
```

```python
Range(
    "How much would you pay for this thing?",
    prepend="$",
    append=".00",
    default=20
).display()
```

```python
question = Range("What's your favorite number?")
question.set_is_valid(False)
question.feedback = "Your favorite number is incorrect."
question.display()
```

```python
question = Range(
    "Kitchen sink",
    prepend=["a", "b", "c "],
    append=[" x", "y", "z"],
    form_text="Here is some form text."
)
question.set_is_valid(True)
question.feedback = "Here is some feedback."
question.display()
```
