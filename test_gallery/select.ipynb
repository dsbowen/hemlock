```python
from hemlock import create_test_app
from hemlock.questions import Select

app = create_test_app()
```

```python
Select(
    "What's your favorite ice cream flavor?",
    ["Chocolate", "Vanilla", "Strawberry"]
).display()
```

```python
Select(
    "Which ice cream flavors do you like?",
    ["Chocolate", "Vanilla", "Strawberry"],
    multiple=True
).display()
```

```python
Select(
    "This question has a default response.",
    [0, 1, 2],
    default=1
).display()
```

```python
Select(
    "This question has multiple defaults.",
    [0, 1, 2],
    multiple=True,
    default=[1, 2]
).display()
```

```python
select = Select(
    "Which survey software is better?",
    ["Hemlock", "Qualtrics"],
    form_text="Hint: There is a correct answer."
)
select.raw_response = "Hemlock"
select.set_is_valid(True)
select.feedback = "Yup!"
select.display()
```

```python
select = Select(
    "Kitchen sink",
    [0, 1, 2],
    default=1,
    floating_label="Floating",
    prepend=["a", "b", "c"],
    append=["x", "y", "z"],
    form_text="Here is some form text."
)
select.set_is_valid(False)
select.feedback = "Here is some feedback"
select.display()
```
