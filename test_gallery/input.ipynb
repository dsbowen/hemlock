```python
from hemlock import create_test_app
from hemlock.questions import Input

app = create_test_app()
```

```python
Input("What's your name?").display()
```

```python
Input("How old are you?", input_tag={"type": "number", "min": 0}).display()
```

```python
Input(
    "How much would you pay for this thing?",
    prepend="$",
    append=".00",
    input_tag={
        "type": "number",
        "min": 0,
        "placeholder": 0
    }
).display()
```

```python
Input(floating_label="Email").display()
```

```python
input = Input("What's your name?")
input.set_is_valid(False)
input.feedback = "Please enter your name."
input.display()
```

```python
input = Input(
    "Kitchen sink",
    default="1992-10-26",
    floating_label="Floating",
    prepend=["a", "b", "c"],
    append=["x", "y", "z"],
    form_text="Here is some form text.",
    input_tag={"type": "date"}
)
input.set_is_valid(True)
input.feedback = "Here is some feedback."
input.display()
```
