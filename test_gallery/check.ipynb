```python
from hemlock import create_test_app
from hemlock.questions import Check

app = create_test_app()
```

```python
Check(
    "What's your favorite ice cream flavor?",
    ["Chocolate", "Vanilla", "Strawberry"]
).display()
```

```python
Check(
    "Which ice cream flavors do you like?",
    ["Chocolate", "Vanilla", "Strawberry"],
    multiple=True
).display()
```

```python
levels_of_agreement = ("Disagree", "Neither agree nor disagree", "Agree")
Check(
    "How much do you agree?",
    [(i, level) for i, level in enumerate(levels_of_agreement)]
).display()
```

```python
Check(
    "This question has disabled choices.",
    ["Enabled", {"value": "Disabled", "disabled": True}]
).display()
```

```python
Check(
    "This question has a default response.",
    [0, 1, 2],
    default=1
).display()
```

```python
Check(
    "This question has multiple defaults.",
    [0, 1, 2],
    multiple=True,
    default=[1, 2]
).display()
```

```python
Check(
    "This check uses inline choices.",
    [0, 1, 2],
    inline=True
).display()
```

```python
Check(
    "This check uses switch choices.",
    [0, 1, 2],
    switch=True,
).display()
```

```python
check = Check(
    "Which survey software is better?",
    ["Hemlock", "Qualtrics"],
    form_text="Hint: There is a correct answer.",
    switch=True,
    inline=True
)
check.raw_response = "Hemlock"
check.set_is_valid(True)
check.feedback = "Yup!"
check.display()
```
