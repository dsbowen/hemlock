```python
from hemlock.app import create_test_app
from hemlock.questions import Label, Input

app = create_test_app()
```

```python
Input("What's your name?").display()
```

```python
Input(
    "How much would you pay for this thing?",
    prepend="$",
    append=".00",
    input_tag={
        "type": "number",
        "placeholder": "0.00",
        "min": 0,
        "max": 100
    }
).display()
```

```python
Input(
    "Enter your email address.",
    floating_label="Email"
).display()
```

```python
input = Input(
    "Enter a dollar amount.",
    prepend=["$", ".00"]
)
input.set_is_valid(False)
input.feedback = "Please enter a valid amount."
input.display()
```

```python
input = Input(
    "Your vanity URL.",
    floating_label="Username",
    append=["@", "example.com"]
)
input.set_is_valid(True)
# note: bootstrap CSS appears not to render feedback with floating labels
input.display()
```
